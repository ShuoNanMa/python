# 5.4 时间序列建模举例
时间序列建模是计量经济学的一个分支，在古扎拉蒂的《计量经济学基础》（第五版）的第21-22章里对其基本概念和预测模型有介绍。![](figures/计量经济学基础.png)

我们在讲课中会引用部分内容。

## 5.4.1 导入数据

首先，我们来看一个数据，它是航空公司从1949年到1960年每个月的旅客人数。

```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
from statsmodels.tsa.stattools import adfuller

data = pd.read_csv('data/AirPassengers.csv', index_col='Month')
data
```

```python
data.index
```

该索引的类型是'object'。为使用时间序列，需要将索引转换为日期时间型：

```python
data.index = pd.to_datetime(data.index)
data.index
```

从它的图像中可以看到，它呈现逐年上升的趋势，同时随季节波动：

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']#百度后：属性总结（三）：plt.rcParams
rcParams['figure.figsize'] = 10, 5
data.plot()
```

# 5.4.2 画图分析

```python
import pandas as pd
passengers = data['Passengers']
plot_data = pd.DataFrame({'旅客数': passengers, '12个月移动平均线': passengers.rolling(window=12).mean()})
plot_data.plot()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(passengers, model="additive")
f = plt.figure(facecolor='white')
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid
ax1 = f.add_subplot(3,1,1)
ax1.plot(trend)#长期趋势
ax2 = f.add_subplot(3,1,2)
ax2.plot(seasonal)#季节趋势
ax3 = f.add_subplot(3,1,3)
ax3.plot(residual)#剩余部分（噪声）
print("trend['1949-01-01':'1950-12-01']:\n", trend['1949-01-01':'1950-12-01'])#看看数据
```

# 5.4.3 平稳性检验

平稳的随机时间序列粗略地讲，一个时间序列，如果均值没有系统的变化（无趋势）、方差没有系统变化，且严格消除了周期性变化，就称之是平稳的。![](figures/弱平稳.png)

**时间序列建模需要事先检验一个序列是否是平稳的。如否，将一个非平稳的序列转换为平稳序列。** 常采用ADF检验（即单位根检验），它的原假设为序列具有单位根，即非平稳。该检验结果的p值是0.99（如下），不能拒绝原假设，即非平稳，与肉眼观察结果一致。

```python
import statsmodels as sm
t = sm.tsa.stattools.adfuller(passengers)  # ADF检验
output=pd.DataFrame(index=['Test Statistic Value', "p-value", "Lags Used", "Number of Observations Used","Critical Value(1%)","Critical Value(5%)","Critical Value(10%)"],columns=['value'])
output['value']['Test Statistic Value'] = t[0]        #t值
output['value']['p-value'] = t[1]                     #p值
output['value']['Lags Used'] = t[2]                   #阶数
output['value']['Number of Observations Used'] = t[3]
output['value']['Critical Value(1%)'] = t[4]['1%']
output['value']['Critical Value(5%)'] = t[4]['5%']
output['value']['Critical Value(10%)'] = t[4]['10%']
output
```

**常用对原序列取log或差分（可以是多阶）来使得变换后的序列是平稳的。**

```python
np.log(passengers) #看看数据
```

```python
np.log(passengers).shift(1) #看看数据
```

```python
passengers_log_diff = (np.log(passengers)-np.log(passengers).shift(1)).dropna() #log序列的一阶差分
passengers_log_diff
```

```python
passengers_log_diff.plot()
```

对该序列再做平稳性检验，如下p值为0.07，在10%的置信度上是平稳的：

```python
t = sm.tsa.stattools.adfuller(passengers_log_diff)  # ADF检验
output=pd.DataFrame(index=['Test Statistic Value', "p-value", "Lags Used", "Number of Observations Used","Critical Value(1%)","Critical Value(5%)","Critical Value(10%)"],columns=['value'])
output['value']['Test Statistic Value'] = t[0]        #t值
output['value']['p-value'] = t[1]                     #p值
output['value']['Lags Used'] = t[2]                   #阶数
output['value']['Number of Observations Used'] = t[3]
output['value']['Critical Value(1%)'] = t[4]['1%']
output['value']['Critical Value(5%)'] = t[4]['5%']
output['value']['Critical Value(10%)'] = t[4]['10%']
output
```

# 5.4.4 确定模型
   采用ARIMA模型（《计量经济学基础》22.2 AR, MA, and ARIMA Modeling of Time Series Data），首先画出自相关和偏相关图，来确定阶数（Table 22.1）：

```python
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
f = plt.figure(facecolor='white')
ax1 = f.add_subplot(211)
plot_acf(passengers_log_diff, lags=31, ax=ax1)
ax2 = f.add_subplot(212)
plot_pacf(passengers_log_diff, lags=31, ax=ax2)
plt.show()
```

观察上图，发现自相关和偏相系数都存在拖尾的特点，并且他们都具有明显的一阶相关性，所以我们设定p=1, q=1。下面就可以使用ARMA模型进行数据拟合了。

```python
from statsmodels.tsa.arima_model import ARMA
model = ARMA(passengers_log_diff, order=(1, 1)) 
result_arma = model.fit( disp=-1, method='css')#拟合
```

# 5.4.5 预测和还原

```python
predict = result_arma.predict()
predict
#如何还原？
```

因为对处理后的数据做模型，我们还需要还原到原数据。做个小练习吧，让我们还原原始数据，并画出它的图。

```python
passengers
```

```python
x = [1, 2, 3, 4]
np.add.accumulate(x)#用到它
```

```python
np.add.accumulate(predict)#由差值转换为累加值
```

```python
np.add.accumulate(predict)+ np.log(passengers['1949-02-01'])#加上起始点的值来修正一下
```

```python
predict_passengers = np.exp(np.add.accumulate(predict)+ np.log(passengers['1949-02-01'])) #log函数的反函数 
```

```python
import matplotlib.pyplot as plt
plt.plot(passengers)
plt.plot(predict_passengers)
plt.legend(['原值', '预测值'])
```
